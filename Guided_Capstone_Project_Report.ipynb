# Problem Statement
The goal of this project was to determine which factors have the greatest influence on adult weekend ski resort ticket prices and to build a machine learning model that could accurately predict those prices. By identifying the features that contribute most to ticket pricing, the analysis can help ski resorts make more informed business and pricing decisions.

The project began by loading a dataset containing information on 330 ski resorts across the United States. Before any analysis could be completed, I cleaned and prepared the data to improve its quality. This included identifying duplicate resort names, correcting inaccurate values, removing columns with too many missing values, filtering unrealistic data, and removing rows with missing ticket prices. To make the analysis more meaningful, I also merged state demographic information, including population and land area, into the dataset. This additional information allowed me to compare resorts within the context of their individual states.

After cleaning the data, I performed exploratory data analysis (EDA) to better understand the relationships between the variables. I created several charts and visualizations to examine the distribution of ski resorts across the country, compare ticket prices by state, and identify trends between resort features and pricing. I also engineered several new features by creating ratios that compared each resort to state-level averages. In addition, I used Principal Component Analysis (PCA) to reduce the complexity of the state-level data while still capturing most of the variation. One of the strongest findings from the analysis was that larger resorts with more skiable terrain, greater vertical drop, more chairlifts, and longer operating seasons generally had higher weekend ticket prices. States such as Colorado and California also tended to have higher average ticket prices than many Midwestern states.

To predict adult weekend ticket prices, I built multiple machine learning models. I first created a Dummy Regressor to establish a baseline, then developed both a Linear Regression model and a Random Forest Regressor. During preprocessing, I handled missing values with median imputation, applied feature selection using SelectKBest, and used GridSearchCV to optimize the models. Performance was evaluated using R-squared, Mean Absolute Error (MAE), Mean Squared Error (MSE), and cross-validation to determine which model generalized best to unseen data.

The Random Forest Regressor proved to be the strongest model. It achieved a cross-validation R-squared of approximately 0.71 and a test MAE of about 9.54, outperforming the Linear Regression model, which had a test MAE of $11.79. The Random Forest also showed consistent performance across both the training and testing datasets, making it the most reliable model for predicting weekend ticket prices.

Finally, I used the model to evaluate pricing scenarios for Big Mountain Resort. The model estimated a ticket price of approximately 95.87, compared to the resort’s current weekend ticket price of $81.00, suggesting there may be an opportunity to increase pricing. I also tested several operational scenarios and found that reducing the number of ski runs would decrease both ticket price and seasonal revenue. In contrast, investing in improvements such as adding ski runs, increasing vertical drop, expanding chairlift capacity, and improving snow-making capability supported higher ticket prices and projected millions of dollars in additional seasonal revenue.

# Pricing Recommendation

Based on the results of the Random Forest model, I would recommend that Big Mountain Resort consider a gradual increase in its adult weekend ticket price while continuing to invest in resort improvements. The analysis suggests that guests place value on features such as additional runs, greater vertical drop, and enhanced infrastructure, making these investments more likely to support higher prices and increased long-term revenue.
